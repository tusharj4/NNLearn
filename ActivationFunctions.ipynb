```python
import numpy as np
import matplotlib.pyplot as plt
```

***Linear Identity Activation Function***

Linear functions are pretty simple. It returns what it gets as input.

*formula:* **$$f(x)=x$$**

*Range:* **{-inf, +inf}**

*derivative:* **$$F'(x)=1$$**

```python
def linear(x):
    y = x
    return y
```

```python
X = np.random.randn(10)
y = np.zeros(10);

for i in range(10):
   y[i:] = linear(X[i:])

plt.plot(X,y,'-ko', markersize = 7, linewidth = 1.5)
plt.title("Linear Identity Activation Function")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

***Binary Step Activation Function***

Binary step function returns value either **0 or 1**.

* It returns '0' if the input is the less then zero
* It returns '1' if the input is greater than zero

***formula:*** $$f(x)=(1 ⟶ if (x>0)) -OR-(0⟶if (x<0))$$

***Range:*** **{0, 1}**

***derivative:*** **$$F'(x)=(0 ⟶ if (x!=0)) -OR-(undefined⟶if (x=0))$$**

```python
def binarystep(x):
  data= [1 if x>0 else 0]
  return np.array(data, dtype=float)
```

```python
values = np.arange(-10, 10, 0.1)
y=[binarystep(i) for i in values]
plt.plot(values,y )
plt.xlabel('x')
plt.ylabel('binarystep(x)')
plt.title('Binary Step Function in Matplotlib')
plt.show()
```

***Sigmoid Activation Function***

*formula:* $$f(x)=1/(1+e^{-x})$$

*Range:* **(0, 1)**

*derivative:* **$$F'(x)=f(x)(1-f(x))$$**

```python
def sigmoid(x):
  return 1.0/(1.0+np.exp(-x))
```

```python
values = np.arange(-10, 10, 0.1)

plt.plot(values, sigmoid(values))
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.title('Sigmoid Function in Matplotlib')
plt.show()
```

***Rectified Linear Activation Function***

*formula:* $$f(x)={0, if (x<0)} -OR
-f(x)={1, if (x>0)}$$

*Range:* **[0,inf)**

*derivative:* **$(0⟶{if x<0}| 1⟶{if x>0}| undefined⟶{if x=0})$**

```python
def ReLU(x):
  data = [max(0,value) for value in x]
  return np.array(data, dtype=float)
```

```python
values = np.arange(-10, 10, 0.1)
y=ReLU(values)
plt.plot(values, y)
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.title('Rectified Linear Function in Matplotlib')
plt.show()
```

***Tangent Hyperbolic Activation Function***

Tanh is another nonlinear activation function. Tanh outputs between -1 and 1. Tanh also suffers from gradient problem near the boundaries just as Sigmoid activation function does.

***formula:*** $$f(x)=(e^{x}-e^{-x})/(e^{x}+e^{-x})$$

***Range:*** **(-1, 1)**

***derivative:*** **$$F'(x)=1-f(x)^2$$**

```python
def tanhi(x):
  return ((np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x)))
```

```python
values = np.arange(-10, 10, 0.1)
y=tanhi(values)
plt.plot(values, y)
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.title('Tanh in Matplotlib')
plt.show()
```

***Softmax Activation Function***

Softmax turns logits, the numeric output of the last linear layer of a multi-class classification neural network into probabilities.

***formula:***$$\sigma(o_i) = \frac{e^{o_i}}{\sum_{j=1}^{n} e^{o_j}}$$

***Range:*** **(0, 1)**


```python
def softmax(x):
  return np.exp(x) / np.sum(np.exp(x), axis=0)
```

```python
values = np.arange(-10, 10, 0.1)
y=softmax(values)
plt.plot(values, y)
plt.xlabel('x')
plt.ylabel('softmax(x)')
plt.title('softmax in Matplotlib')
plt.show()
```
