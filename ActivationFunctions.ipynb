```python
import numpy as np
import matplotlib.pyplot as plt
```

***Linear Identity Activation Function***

*formula:* **f(x)=x**

*Range:* **{-inf, +inf}**

*derivative:* **F'(x)=1**

```python
def linear(x):
    y = x
    return y
```

```python
X = np.random.randn(10)
y = np.zeros(10);

for i in range(10):
   y[i:] = linear(X[i:])

plt.plot(X,y,'-ko', markersize = 7, linewidth = 1.5)
plt.title("Linear Identity Activation Function")
# plt.xlim([-1.5, 2.5])
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

***Sigmoid Activation Function***

*formula:* $f(x)=1/(1+e^{-x})$

*Range:* **(0, 1)**

*derivative:* **$F'(x)=f(x)(1-f(x))$**

```python
def sigmoid(x):
  return 1.0/(1.0+np.exp(-x))
```

```python
values = np.arange(-10, 10, 0.1)

plt.plot(values, sigmoid(values))
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.title('Sigmoid Function in Matplotlib')
plt.show()
```

***Rectified Linear Activation Function***

*formula:* $f(x)={0, if (x<0)}$ OR
$f(x)={1, if (x>0)}$

*Range:* **[0,inf)**

*derivative:* **$(0⟶{if x<0}| 1⟶{if x>0}| undefined⟶{if x=0})$**

```python
def ReLU(x):
  data = [max(0,value) for value in x]
  return np.array(data, dtype=float)
```

```python
values = np.arange(-10, 10, 0.1)
y=ReLU(values)
plt.plot(values, y)
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.title('Rectified Linear Function in Matplotlib')
plt.show()
```

***Tangent Hyperbolic Activation Function***

*formula:* $f(x)=(e^{x}-e^{-x})/(e^{x}+e^{-x})$

*Range:* **(-1, 1)**

*derivative:* **$F'(x)=1-f(x)^2$**

```python
def tanhi(x):
  return ((np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x)))
```

```python
values = np.arange(-10, 10, 0.1)
y=tanhi(values)
plt.plot(values, y)
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.title('Tanh in Matplotlib')
plt.show()
```
